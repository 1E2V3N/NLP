```python
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
```

```python
data=pd.read_csv("trumptweets.csv")
```

```python
data.head()
```

```python
col=data['content']
```

```python
col_data=col.iloc[:20]
```

```python
col_data
```

```python
sentence = ' '.join(col_data)
```

```python
sentence
```

```python
# remove stopwords function
def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    return filtered_text
```

```python
words=remove_stopwords(sentence)
```

##### words

```python
for i in range(len(words)):
    words[i] = words[i].lower()
    words[i] = re.sub(r'\W', ' ', words[i])
    words[i] = re.sub(r'\s+', ' ', words[i])
```

```python
words
```

**Stemming**

```python
ps=PorterStemmer()
```

```python
stemmed=[]
```

```python
for word in words:
    stemmed_word=ps.stem(word)
    stemmed.append(stemmed_word)
```

```python
print(stemmed)
```

```python
compare=pd.DataFrame(data={"tokenized":words,"stemmed":stemmed})
compare.iloc[0:25,]
```

**Lemmatization**

```python
nltk.download('wordnet')
```

```python
wn=WordNetLemmatizer()
```

```python
lemmatized=[]
```

```python
for word in words:
    lemmatized_word=wn.lemmatize(word)
    lemmatized.append(lemmatized_word)
```

```python
print(lemmatized)
```

```python
compared=pd.DataFrame(data={"tokenized":words,"stemmed":stemmed,"lemmatized":lemmatized})
compared.iloc[0:50,0:3]
```
